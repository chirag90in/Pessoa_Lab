```python
import pandas as pd
```

```python
#machine = 'centos-7'
machine = 'centos-7-02'

cov_df = pd.read_csv('../01-uncon_v_con_shock_MR_ROI24_with_covariates/uncon_v_con_shock_MR_ROI24final_with_covariates.txt',sep='\t')
df = pd.read_csv(machine+'/uncon_v_con_lInsula_shock.txt',sep='\t')
print('cov shape:', cov_df.shape)
print('df shape:', df.shape)
cov_df.head()
```

```python
cov_df = cov_df.drop(['ROI','Y'],axis=1).drop_duplicates().reset_index(drop=True)
```

```python
cov_df.head()
```

```python
df.head()
```

```python
df = df.merge(right=cov_df)
df.head()
```

```python
df.to_csv(machine+'/uncon_v_con_lInsula_shock.txt',sep='\t',index=False,float_format='%.4f')
```
