```python
from scipy.stats import gamma

import math
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# **Gamma Distribution**
$$\gamma(\alpha,\sigma^{2},\lambda)$$

- $\alpha$: mean of Gaussian component
- $\sigma^{2} > 0$: variance of Gaussian component
- $\lambda > 0$: rate of exponential component

## **Probability and Cumulative Dirstribution plots**
- $\gamma(\alpha=2, \lambda =0.1)$ 

```python
alpha, rate = 2, 0.1
beta = 1/rate
avrg = alpha*beta
std_dv = math.sqrt(alpha*beta**2)
# Draw the gamma distribution
rng = np.arange(0,avrg+5*std_dv,0.01)

x = 3

y_pdf = gamma.pdf(rng,a=alpha,scale=beta)# scale = 1/rate
y_cdf = gamma.cdf(rng,a=alpha,scale=beta)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(rng,y_pdf,'k')
idx = rng < 3
plt.fill_between(rng[idx],y_pdf[idx],color='red',alpha=0.5)
plt.annotate(s='Area < %d = %.3f' %(x,gamma.pdf(x,a=alpha,scale=beta)),
             xy = (x,0.015),
             xytext=(25,0.025),
             arrowprops={'color':'red','width':1})
_=plt.ylim(0,np.max(y_pdf)+0.01)
_=plt.title('PDF')

plt.subplot(1,2,2)
plt.plot(rng,y_cdf,'k')
plt.fill_between(rng[idx],y_cdf[idx],color='red',alpha=0.5)

plt.annotate(s='Area < %d = %.3f' %(x,gamma.cdf(x,a=alpha,scale=beta)),
             xy = (x,0.015),
             xytext=(25,0.025),
             arrowprops={'color':'red','width':1})

_=plt.ylim(0,np.max(y_cdf)+0.01)
_=plt.title('CDF')

# Creating inset to zoom in on the shaded region
plt.axes([0.7,0.5,0.15,0.3])
plt.plot(rng,y_cdf,'k')
plt.fill_between(rng[idx],y_cdf[idx],color='red',alpha=0.5)

_=plt.ylim(0,0.2)
_=plt.xlim(0,6)

print('Standard devation = %.2f'%(std_dv))
```
